```python
import pandas as pd
import seaborn as sns
```

```python
df_1 = pd.read_csv("runtime_old.csv")
df_1["old"] = True

df_2 = pd.read_csv("runtime1.csv")
df_2["old"] = False

df_3 = pd.read_csv("runtime.csv")
df_3["old"] = False

df = pd.concat((df_1, df_2, df_3), axis=0, ignore_index=True)
df["job"] = df["job"].map(lambda x: str(x))

sns.relplot(df, x="n", y="time", hue="job", kind="line")
```
